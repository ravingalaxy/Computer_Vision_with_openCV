```
gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Convert from cv's BRG default color order to RGB
print(cv2_imshow(gray_image))
```

```
# Converting image to HSV format

new_image = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

# plotting the HSV image

print(cv2_imshow(new_image))
```

```
smaller_image_size = cv2.resize(img,(250,200))
print(cv2_imshow(smaller_image_size))
```

```
ret, thresh_binary = cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)
ret, thresh_binary_inv = cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY_INV)
ret, thresh_trunc = cv2.threshold(gray_image,127,255,cv2.THRESH_TRUNC)
ret, thresh_tozero = cv2.threshold(gray_image,127,255,cv2.THRESH_TOZERO)
ret, thresh_tozero_inv = cv2.threshold(gray_image,127,255,cv2.THRESH_TOZERO_INV)
```

```
# Displaying the different thresholding styles using openCV
names = ['Original Image','THRESH_BINARY','THRESH_BINARY_INV','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']
images = gray_image,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv

for i in range(6):

  plt.subplot(2,3,i+1), plt.imshow(images[i],'gray')

  plt.title(names[i])

  plt.xticks([]),plt.yticks([])
```

```
face_cascade =  cv2.CascadeClassifier("/content/drive/My Drive/haarcascade_frontalface_default.xml")
```

```
faces = face_cascade.detectMultiScale(gray_image,1.1,4)
# Draw rectangle around the faces
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w, y+h),(255,0,0),2)
#Display the output
cv2_imshow(img)
cv2.waitKey()
```

```
multiface_img = cv2.imread('/content/drive/My Drive/multiface.jpg')
print(cv2_imshow(multiface_img))
```

```

gray_multiface_image = cv2.cvtColor(multiface_img,cv2.COLOR_BGR2GRAY) # Convert from cv's BRG default color order to RGB

faces = face_cascade.detectMultiScale(gray_multiface_image,1.1,4)
# Draw rectangle around the faces
for (x,y,w,h) in faces:
  cv2.rectangle(multiface_img,(x,y),(x+w, y+h),(255,0,0),2)
#Display the output
cv2_imshow(multiface_img)
cv2.waitKey()
```
